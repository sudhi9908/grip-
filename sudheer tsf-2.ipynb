## Task  2 - To Explore Supervised Machine Learning


In this regression task we will predict the percentage of marks that a student is expected to score based upon the number of hours they studied. This is a simple linear regression task as it involves just two variables.


Data can be found at http://bit.ly/w-data

### Question is : What will be predicted score if a student study for 9.25 hrs in a day

#### Importing libraries¶


```python
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt  
%matplotlib inline
```

#### Reading data

```python
url="http://bit.ly/w-data"
df=pd.read_csv(url)
print("Data imported successfully")

df.head(20)
```

###  Plotting data¶

```python
plt.scatter(df.Hours, df.Scores, color="g") 
plt.title('Hours vs Percentage')
plt.xlabel('Hours Studied')  
plt.ylabel('Percentage Score')  
plt.show()
```


### Preparing the data

```python
x = df.iloc[:, :-1].values
y = df.iloc[:, 1].values
```

```python
from sklearn.model_selection import train_test_split  
x_train, x_test, y_train, y_test = train_test_split(x, y, 
                            test_size=0.2, random_state=0)
```

### Training the algorithm¶


```python
from sklearn.linear_model import LinearRegression  
reg = LinearRegression()  
reg.fit(x_train, y_train) 

print("Training completed.")
```

```python
print(x_train)
```

```python
print(y_train)
```

### plotting

```python
line = reg.coef_*x+reg.intercept_
plt.scatter(x, y)
plt.plot(x, line)
plt.show()
```

### Testing and predictions

```python
print(x_test) #testing the data
predy = reg.predict(x_test) #predicting the scores
```

```python
df = pd.DataFrame({'Actual': y_test, 'Predicted': predy})  
print(df)
```

```python
predy=reg.predict(x_test)
predy
```

### Evaluating accuracy of the model

```python
from sklearn import metrics
metrics.r2_score(y_test,predy)
```

Implies that 94% of the model is of good fit¶


### Error prediction¶


```python
from sklearn import metrics  
print('Mean Absolute Error:', 
      metrics.mean_absolute_error(y_test, predy))
```

### Predicting final score

```python
y1_pred=reg.predict([[9.25]])
print('predicted score :',y1_pred)
```
